## NetworkX

```python
# Extração de Embedings

model = SentenceTransformer('all-MiniLM-L6-v2')  # leve e bom para similaridade
df['embedding'] = df['message'].apply(lambda x: model.encode(x, convert_to_tensor=True))
```

```python
# Inicializa o grafo
G = nx.Graph()

# Agrupa os dados
grouped = df.groupby(['id_group_anonymous'])
threshold = 0.75

for group_id, group_df in grouped:
    users = group_df['id_member_anonymous'].unique()
    embeddings = group_df["embedding"].tolist()
    n = len(users)

    for i in range(n):
        for j in range(i+1, n):
            sim = util.cos_sim(embeddings[i], embeddings[j]).item()
            if sim >= threshold:
                u1, u2 = users[i], users[j]

                # Adiciona os nós (se não existirem)
                G.add_node(u1)
                G.add_node(u2)

                # Se a aresta já existe, incrementa o peso
                if G.has_edge(u1, u2):
                    G[u1][u2]['weight'] += 1
                else:
                    G.add_edge(u1, u2, weight=1)
```

```python
# Visualizar no PyVis
net = Network(notebook=True)
net.from_nx(G)
# Mostra no notebook
net.show('grafos/grafo_geral_networkx.html')

# Abre no navegador
webbrowser.open('grafos/grafo_geral_networkx.html')
```

```python
# Filtra mensagens virais
viral_df = df[df['viral'] == True]

# Agrupa por grupo e conteúdo da mensagem
grouped = viral_df.groupby(['id_group_anonymous', 'text_content_anonymous'])

# Inicializa o grafo
GV = nx.Graph()

# Cria relacionamentos entre usuários que compartilharam a mesma mensagem viral
for (group_id, message), group_df in grouped:
    users = group_df['id_member_anonymous'].unique()
    n = len(users)

    for i in range(n):
        for j in range(i + 1, n):
            u1, u2 = users[i], users[j]

            # Adiciona nós (caso ainda não existam)
            GV.add_node(u1)
            GV.add_node(u2)

            # Incrementa peso se já existe a aresta
            if GV.has_edge(u1, u2):
                GV[u1][u2]['weight'] += 1
            else:
                GV.add_edge(u1, u2, weight=1)

# print(f"Grafo criado com {G.number_of_nodes()} nós e {G.number_of_edges()} arestas.")
```

```python
# Visualizar no PyVis
net = Network(notebook=True)
net.from_nx(GV)
# Mostra no notebook
net.show('grafos/grafo_viral_networkx.html')

# Abre no navegador
webbrowser.open('grafos/grafo_viral_networkx.html')
```

![Texto alternativo](imagens/graph_viral_networkx.png)

```python
# Filtrar somente mensagens de desinformação
misinfo_df = df[df['is_misinformation'] == True]

# Agrupar por grupo e conteúdo da mensagem
grouped = misinfo_df.groupby(['id_group_anonymous', 'text_content_anonymous'])

# Criar grafo
GM = nx.Graph()

# Criar relacionamentos entre usuários que compartilharam a mesma mensagem de desinformação
for (group_id, message), group_df in grouped:
    users = group_df['id_member_anonymous'].unique()
    n = len(users)

    for i in range(n):
        for j in range(i + 1, n):
            u1, u2 = users[i], users[j]

            # Adicionar nós (se ainda não existirem)
            GM.add_node(u1)
            GM.add_node(u2)

            # Se já existe uma aresta, incrementa o peso
            if GM.has_edge(u1, u2):
                GM[u1][u2]['weight'] += 1
            else:
                GM.add_edge(u1, u2, weight=1)

# print(f"Grafo criado com {G.number_of_nodes()} nós e {G.number_of_edges()} arestas.")
```

#### Verifique se existem grupos isolados e clusters de usuários fortemente conectados. Isso pode ocorrer caso existam usuários engajados que possuem participação ativa em diversos grupos. 

##### Grafo Geral

```python
# Verificar grupos isolados (componentes desconectados)
# Componentes conectados
components = list(nx.connected_components(G))

print(f"Quantidade de grupos isolados: {len(components)}")
print("Tamanho dos grupos:")
for i, comp in enumerate(sorted(components, key=len, reverse=True)[:100], 1):
    print(f"Grupo {i}: {len(comp)} nós")
```

```python
# Detectar clusters de usuários fortemente conectados
communities = list(greedy_modularity_communities(G))
print(f"Comunidades detectadas: {len(communities)}")

for i, c in enumerate(sorted(communities, key=len, reverse=True)[:100], 1):
    print(f"Comunidade {i}: {len(c)} nós")
```

```python
# Giant Component (maior componente conexo)
# Extrai o maior componente conexo do grafo Geral a partir da lista de componentes
components = list(nx.connected_components(G))
giant = max(components, key=len)
G_giant = G.subgraph(giant).copy()
print(f"Componentes encontrados: {len(components)}")
print(f"Tamanho do Giant Component: {len(G_giant.nodes())} nós e {len(G_giant.edges())} arestas")
```

```python
# Density (densidade)
# Calcula a densidade do grafo Geral indicando a quantidade de conexões existentes em relação ao número máximo possível de conexões
density = nx.density(G)
print(f"Densidade: {density:.4f}")
```

```python
# Diameter (diâmetro)
# Só pode ser calculado em componentes conexos do grafo geral indicando a maior distância mínima entre dois nós
diameter = nx.diameter(G_giant)
print(f"Diâmetro (maior distância mínima entre dois nós): {diameter}")
```

```python
# Degree Centrality
# Mede a importância de um nó baseado no número de conexões que ele possui
degree_centrality = nx.degree_centrality(G)
top_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 Degree Centrality:", top_degree)
```

```python
# Clustering Coefficient
# Mede a tendência dos nós de formarem grupos ou clusters
clustering = nx.average_clustering(G)
print(f"Coeficiente médio de agrupamento (Clustering): {clustering:.4f}")
```

```python
# Closeness Centrality
# Mede a importância de um nó baseado na sua proximidade com todos os outros nós do grafo
closeness = nx.closeness_centrality(G)
top_closeness = sorted(closeness.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 Closeness Centrality:", top_closeness)
```

```python
# Betweenness Centrality
# Mede a importância de um nó baseado na quantidade de vezes que ele atua como ponte ao longo do caminho mais curto entre dois outros nós
betweenness = nx.betweenness_centrality(G)
top_betweenness = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 Betweenness Centrality:", top_betweenness)
```

#### Implemente algoritmos de predição de ligações utilizando as métricas a seguir: 

##### Preferential attachment

```python
# Preferential Attachment
# Mede a probabilidade de novos nós se conectarem a nós já bem conectados
pa = list(nx.preferential_attachment(G))

df_pa = pd.DataFrame(pa, columns=['u', 'v', 'preferential_attachment'])
print(df_pa.head(10))
```

##### Common neighbors

```python
# Common Neighbors
# Mede a quantidade de vizinhos em comum entre dois nós
cn = [(u, v, len(list(nx.common_neighbors(G, u, v)))) for u, v in nx.non_edges(G)]

df_cn = pd.DataFrame(cn, columns=['u', 'v', 'common_neighbors'])
print(df_cn.head(10))
```

##### Jaccard 

```python
# Jaccard Coefficient
# Mede a similaridade entre dois nós baseado na proporção de vizinhos em comum
jc = list(nx.jaccard_coefficient(G))

df_jc = pd.DataFrame(jc, columns=['u', 'v', 'jaccard'])
print(df_jc.head(10))
```

##### Adamic-Adar

```python
# Adamic-Adar Index
# Mede a similaridade entre dois nós dando mais peso a vizinhos menos conectados
aa = list(nx.adamic_adar_index(G))

df_aa = pd.DataFrame(aa, columns=['u', 'v', 'adamic_adar'])
print(df_aa.head(10))
```

##### Graph distance

```python
# Graph Distance (usaremos distância inversa)
# Mede a proximidade entre dois nós baseado na distância mínima entre eles
shortest_paths = dict(nx.all_pairs_shortest_path_length(G))
gd = []
for u, v in nx.non_edges(G):
    try:
        dist = shortest_paths[u][v]
        gd.append((u, v, 1/dist))  # quanto menor a distância, maior a probabilidade
    except KeyError:
        gd.append((u, v, 0))  # nós desconectados

df_gd = pd.DataFrame(gd, columns=['u', 'v', 'graph_distance'])
print(df_gd.head(10))
```

##### Katz

```python
# Katz Index (propagação global)
# Mede a influência de um nó baseado em todas as suas conexões, diretas e indiretas
# α é um fator de atenuação que reduz a influência de caminhos mais longos;
# parâmetro β deve ser < 1/λ_max (autovalor máximo da matriz de adjacência)
katz = nx.katz_centrality_numpy(G, alpha=0.005, beta=1.0)

# Gerar todos os pares possíveis que ainda não têm aresta
non_edges = list(nx.non_edges(G))

# Criar DataFrame com as pontuações Katz combinadas
df_katz = pd.DataFrame([
    (u, v, katz[u] + katz[v])  # ou katz[u] * katz[v]
    for u, v in non_edges
], columns=['u', 'v', 'katz_score'])

# Visualizar os 10 pares mais prováveis segundo Katz
df_katz_sorted = df_katz.sort_values('katz_score', ascending=False)
print(df_katz_sorted.head(10))
```
