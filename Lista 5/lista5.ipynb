## Ataques de Flooding

### Criar granularidade por minuto (Para medir quantas mensagens cada usuário envia por minuto)

```python
# Mensagens por minuto
df['minute'] = df['date_message'].dt.floor('min')

# Calcula mensagens por minuto por usuário
msgs_per_minute = (
    df.groupby(['id_member_anonymous', 'minute'])['id_message']
      .count()
      .rename('msgs_per_min')
      .reset_index()
)
```

### Calcular estatísticas por usuário (Assim descobrimos o “normal” de cada pessoa)

```python
# Uso de estatísticas descritivas para cada usuário
user_stats = (
    msgs_per_minute.groupby('id_member_anonymous')['msgs_per_min']
    .agg(['mean', 'std', 'max', 'median'])
    .reset_index()
)
```

### Definir limiar automático para flooding

```python
# Cálculo de thresholds para detecção de picos
thresholds = (
    user_stats.assign(
        threshold=lambda x: x['mean'] + 3 * x['std']
    )[["id_member_anonymous", "threshold"]]
)
```

### Cruzar thresholds com as contagens por minuto (Assim detectamos minutos anômalos)

```python
# Junta stats de cada user com cada minuto
flood_df = msgs_per_minute.merge(
    thresholds,
    on='id_member_anonymous',
    how='left'
)

# flag de flooding
flood_df['is_flood'] = flood_df['msgs_per_min'] > flood_df['threshold']
```

### Trazer essa informação de volta ao DataFrame original

```python
# Junta info de flooding de volta ao df original
df = df.merge(
    flood_df[['id_member_anonymous', 'minute', 'is_flood']],
    on=['id_member_anonymous', 'minute'],
    how='left'
)
```

### Burst temporal muito acelerado (Diferença entre mensagens consecutivas do usuário)

```python
# Detecção de rajadas de mensagens (bursts)
df = df.sort_values(['id_member_anonymous', 'date_message'])

# Tempo entre mensagens consecutivas
df['delta'] = (
    df.groupby('id_member_anonymous')['date_message']
    .diff()
    .dt.total_seconds()
)

# Mensagens enviadas em menos de 5 segundos
df['fast_msg'] = df['delta'] < 5   

# Flag de rajada: mais de 5 mensagens em menos de 10 mensagens são verdadeiras
df['burst_flag'] = (
    df.groupby('id_member_anonymous')['fast_msg']
      .transform(lambda x: x.rolling(10).sum() > 5)
)
```

### Regra baseada na entropia (Mensagens muito repetitivas = suspeita)

```python
# Cálculo de entropia das mensagens
def entropy(text):
    # Verifica se o texto está vazio
    if not text:
        return 0
    # Calcula a frequência de cada caractere
    counts = Counter(text)
    # Calcula o tamanho total do texto
    total = len(text)
    # Calcula a entropia usando a fórmula de Shannon
    return -sum((c/total) * math.log2(c/total) for c in counts.values())

# Aplica a função de entropia às mensagens
df['entropy'] = df['message'].fillna('').apply(entropy)

# Detecta mensagens com entropia anormalmente baixa usando IQR
Q1 = df['entropy'].quantile(0.25)
Q3 = df['entropy'].quantile(0.75)
# Calcula o intervalo interquartil (IQR)
IQR = Q3 - Q1

# Define o limite inferior para detectar outliers
low_threshold  = Q1 - 1.5 * IQR

# Flag de entropia baixa
df['entropy_flag'] = df['entropy'] < low_threshold
```

### Combinação final para classificar flooding

```python
# Combina todas as flags para detectar flooding
df['flooding_detected'] = (
    df['is_flood'].fillna(False)
    | df['burst_flag'].fillna(False)
    | df['entropy_flag'].fillna(False)
)
```

```python
# Resultados finais
df_flooding = df[df['flooding_detected'] == True]
print(f"Número de mensagens com flooding detectado: {len(df_flooding)}")
print("\n")
print(df_flooding[['id_group_anonymous','id_member_anonymous','date_message','flooding_detected']])
```

### O que essa estratégia detecta:
* envios muito rápidos (milissegundos / segundos)
* altos volumes por minuto
* spam repetitivo (emoji, letras repetidas)
* explosões de mensagens em sequência
* comportamento anormal por usuário
* picos acima da média individual (não geral)

## Ataques de Sequestro de Grupos

### Contagem de mensagens por grupo por usuário

```python
msg_counts = (
    df.groupby(['id_group_anonymous', 'id_member_anonymous'])
      ['id_message']
      .count()
      .rename('msgs_user_group')
      .reset_index()
)
```

### Calcular proporção de participação no grupo

```python
group_totals = msg_counts.groupby('id_group_anonymous')['msgs_user_group'].sum().rename('group_total')

msg_counts = msg_counts.merge(group_totals, on='id_group_anonymous')
msg_counts['participation'] = msg_counts['msgs_user_group'] / msg_counts['group_total']
```

#### Para cada grupo, qual % de mensagens cada usuário envia?
#### Se um usuário tem > 60% do grupo, já é suspeito.
#### Se ele tem > 80%, quase sempre é sequestro.    

### Detectar dominância REPENTINA usando ordenação temporal

#### Foi Criado atividade diária para cada usuário

```python
df['day'] = df['date_message'].dt.date

daily = (
    df.groupby(['id_group_anonymous','id_member_anonymous','day'])
      ['id_message']
      .count()
      .rename('daily_msgs')
      .reset_index()
)
```

#### Agora detectar anomalia temporal

```python
# mean/std individual
user_day_stats = (
    daily.groupby(['id_group_anonymous','id_member_anonymous'])['daily_msgs']
         .agg(['mean','std','max'])
         .reset_index()
)

# threshold = mean + 3*std
user_day_stats['threshold'] = user_day_stats['mean'] + 3 * user_day_stats['std']

# juntar de volta
daily = daily.merge(user_day_stats, on=['id_group_anonymous','id_member_anonymous'])
daily['dominance_spike'] = daily['daily_msgs'] > daily['threshold']
```

#### Se um usuário nunca falou muito e de repente explode, marca suspeita.

### Identificar takeover (tomada de controle) por queda de diversidade

#### Se um grupo tem muitos usuários ativos e de repente cai para quase 1

```python
# contar usuários únicos por dia por grupo
unique_users = (
    df.groupby(['id_group_anonymous','day'])['id_member_anonymous']
      .nunique()
      .rename('active_users')
      .reset_index()
)
```

```python
unique_users['diversity_drop'] = (
    unique_users['active_users'].shift(1) - unique_users['active_users'] > 5
)
```

#### Se o grupo tinha 10 pessoas falando e passa a ter 1–2, esse é um sinal claro.

### Entropia do conteúdo por dia do grupo (mudança brusca no tema)

```python

def entropy(text):
    if not text:
        return 0
    c = Counter(text)
    total = len(text)
    return -sum((v/total) * math.log2(v/total) for v in c.values())

df['entropy'] = df['message'].fillna("").apply(entropy)

daily_group_entropy = (
    df.groupby(['id_group_anonymous','day'])['entropy']
      .mean()
      .rename('mean_entropy')
      .reset_index()
)
```

#### Sinal de sequestro

```python
daily_group_entropy['entropy_change'] = (
    daily_group_entropy['mean_entropy'].diff().abs() > 2.0
)
```

### Combinação final — Detecção de Sequestro

```python
# juntar tudo
takeover = df.copy()

takeover = takeover.merge(
    msg_counts[['id_group_anonymous','id_member_anonymous','participation']],
    on=['id_group_anonymous','id_member_anonymous'],
    how='left'
)

takeover = takeover.merge(
    daily[['id_group_anonymous','id_member_anonymous','day','dominance_spike']],
    on=['id_group_anonymous','id_member_anonymous','day'],
    how='left'
)

takeover = takeover.merge(
    unique_users[['id_group_anonymous','day','diversity_drop']],
    on=['id_group_anonymous','day'],
    how='left'
)

takeover = takeover.merge(
    daily_group_entropy[['id_group_anonymous','day','entropy_change']],
    on=['id_group_anonymous','day'],
    how='left'
)
```

```python
takeover['group_hijacking'] = (
    (takeover['participation'] > 0.60) |        # dominância
    (takeover['dominance_spike'] == True) |     # mudança brusca
    (takeover['diversity_drop'] == True) |      # queda de diversidade
    (takeover['entropy_change'] == True)        # mudança de tema
)

df_hijacked = takeover[takeover['group_hijacking'] == True]
print(f"Número de mensagens em grupos com possível sequestro: {len(df_hijacked)}")
print("\n")
print(df_hijacked[['id_group_anonymous','id_member_anonymous','date_message','group_hijacking']])
```

### O que essa estratégia detecta:
* Mudança forte de tema ou conteúdo (mudança brusca no tema).
* Detecta queda brusca
* Detecta anomalia temporal
* Calcula proporção de participação no grupo
