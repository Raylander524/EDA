```python
# Extração de Embedings

model = SentenceTransformer('all-MiniLM-L6-v2')  # leve e bom para similaridade
df['embedding'] = df['message'].apply(lambda x: model.encode(x, convert_to_tensor=True))
```

```python
# Inicializa o grafo
G = nx.Graph()

# Agrupa os dados
grouped = df.groupby(['id_group_anonymous'])
threshold = 0.75

for group_id, group_df in grouped:
    # if len(group_df) > 50:
    #     group_df = group_df.sample(n=50, random_state=42)

    users = group_df['id_member_anonymous'].unique()
    embeddings = group_df["embedding"].tolist()
    n = len(users)

    for i in range(n):
        for j in range(i+1, n):
            sim = util.cos_sim(embeddings[i], embeddings[j]).item()
            if sim >= threshold:
                u1, u2 = users[i], users[j]

                # Adiciona os nós (se não existirem)
                G.add_node(u1)
                G.add_node(u2)

                # Se a aresta já existe, incrementa o peso
                if G.has_edge(u1, u2):
                    G[u1][u2]['weight'] += 1
                else:
                    G.add_edge(u1, u2, weight=1)
```

```python
# Visualizar no PyVis
net = Network(notebook=True)
net.from_nx(G)
# Mostra no notebook
net.show('grafos/grafo_geral_networkx.html')

# Abre no navegador
webbrowser.open('grafos/grafo_geral_networkx.html')
```

```python
node_to_group = (
    df[['id_member_anonymous', 'id_group_anonymous']]
    .drop_duplicates()
    .set_index('id_member_anonymous')['id_group_anonymous']
    .to_dict()
    )
```

```python
results = []
partitions = {}

""" Avalia o quanto as comunidades coincidem com os grupos reais."""
def avaliar_coerencia_com_grupos(communities, node_to_group):
    # Cria uma lista para guardar os resultados
    resultados = []
    # Percorre cada comunidade detectada
    for idx, com in enumerate(communities):
        # Recupera os grupos originais dos membros dessa comunidade
        grupos = [node_to_group.get(node) for node in com if node in node_to_group]
        # Pula comunidades sem grupos válidos
        if not grupos:
            continue
        # Encontra o grupo mais comum dentro da comunidade
        mais_comum, freq = Counter(grupos).most_common(1)[0]
        # Calcula a coerência porcentual
        coerencia = freq / len(grupos)
        # Adiciona os resultados à lista
        resultados.append({
            "comunidade_id": idx,
            "grupo_predominante": mais_comum,
            "coerencia_percentual": coerencia * 100,
            "tamanho_comunidade": len(com)
        })
    return resultados

""" Retorna coerência média ponderada (purity) de todas as comunidades."""
def compute_purity_stats(communities, node_to_group):
    # Calcula coerência por comunidade:
    coerencias = avaliar_coerencia_com_grupos(communities, node_to_group)
    # Se não houver coerências calculadas, retorna 0.0
    if not coerencias:
        return 0.0
    # Converte para DataFrame para facilitar cálculo
    dfc = pd.DataFrame(coerencias)
    # Calcula a média ponderada considerando o tamanho da comunidade
    weighted_avg = (dfc["coerencia_percentual"] * dfc["tamanho_comunidade"]).sum() / dfc["tamanho_comunidade"].sum()
    return weighted_avg / 100.0  # volta para escala 0–1

""" Para cada comunidade, calcula a pureza = fração de nós pertencentes ao grupo de telegramas mais comum."""
def compute_purity_per_community(partition_sets, node_to_group):
    purities = []
    counts = []
    # Percorre cada comunidade
    for community in partition_sets:
        # Recupera os grupos dos nós na comunidade
        groups = [node_to_group.get(n) for n in community]
        # trata None como sua própria categoria se o mapeamento estiver ausente
        c = Counter(groups)
        # Recupera o grupo mais comum e sua contagem
        most_common_count = c.most_common(1)[0][1] if c else 0
        # Pega o tamanho da comunidade
        size = len(community)
        # Calcula a pureza
        purity = most_common_count / size if size > 0 else 0.0
        # Adiciona à lista de purezas
        purities.append(purity)
        # Adiciona o tamanho à lista de contagens
        counts.append(size)
    # retornar pureza por comunidade, tamanhos e pureza média ponderada
    weighted_avg = sum(p * s for p, s in zip(purities, counts)) / sum(counts) if sum(counts) > 0 else 0.0
    return purities, counts, weighted_avg

"""
Calcule a modularidade e a qualidade da partição (cobertura, desempenho) para uma partição.

partition_sets: iterável de conjuntos de nós
Retorna um dicionário com modularidade, cobertura e desempenho
"""
def evaluate_partition(G, partition_sets, weight='weight'):
    # A modularidade espera que as comunidades sejam vistas como listas de conjuntos.
    try:
        # Tenta calcular a modularidade
        mod = q.modularity(G, partition_sets, weight=weight)
    except Exception:
        # Se falhar, define como NaN
        mod = float('nan')
    try:
        # Tenta calcular a qualidade da partição
        coverage, performance = q.partition_quality(G, partition_sets)
    except Exception:
        # Se falhar, define como NaN
        coverage, performance = float('nan'), float('nan')
    return {'modularity': mod, 'coverage': coverage, 'performance': performance}

""" Salvar os resultados de um método de detecção de comunidades."""
def save_result(name, partition_sets, params=None, weight='weight', node_to_group=None):
    # Cria uma chave única para o método e seus parâmetros
    key = f"{name}|{str(params) if params else ''}"
    # Padronização da partição
    partition_sets = [set(c) for c in partition_sets]
    # Cálculo das métricas estruturais
    metrics = evaluate_partition(G, partition_sets, weight=weight)
    # Cálculo da pureza por comunidade
    purities, sizes, weighted_purity = compute_purity_per_community(partition_sets, node_to_group)
    # Cálculo da pureza média global
    avg_purity = compute_purity_stats(partition_sets, node_to_group)
    # Criação de um registro de resultado
    row = {
        'method': name,
        'params': params,
        'n_communities': len(partition_sets),
        'modularity': metrics['modularity'],
        'coverage': metrics['coverage'],
        'performance': metrics['performance'],
        'avg_telegram_purity_weighted': weighted_purity,
        'purities_per_community': purities,
        'sizes_per_community': sizes,
        'avg_group_coherence': avg_purity * 100,  # em %
    }
    # Armazenamento dos resultados
    results.append(row)
    partitions[key] = partition_sets

"""
Execute um conjunto de algoritmos de detecção de comunidades em G.
Retorna um DataFrame do pandas com uma linha por partição encontrada e um dicionário com as partições detalhadas.
"""
def run_all_community_methods(GF, node_to_group, weight='weight', max_k_try=8):
    # 
    n = GF.number_of_nodes()

    # 1) Girvan-Newman: iterar e coletar partições para diferentes números de comunidades
    try:
        gn = girvan_newman(GF)
        # tente extrair até max_k_try níveis distintos
        for i, communities in zip(range(max_k_try), gn):
            # Converte cada comunidade em set comum, para evitar problemas em funções subsequentes
            part = [set(c) for c in communities]
            # Salva os resultados do algoritmo
            save_result('Girvan-Newman', part, params={'level': i+1}, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('Girvan-Newman failed:', e)

    # 2) Edge Betweenness Partition (divisive) - tente números diferentes
    try:
        # Testa vários valores de k, que representa o número desejado de comunidades a serem formadas.
        for k in range(2, min(max_k_try+1, n)):
            # Executa o algoritmo com vários tamanhos de k
            part = edge_betweenness_partition(GF, k, weight=weight)
            # Salva os resultados do algoritmo
            save_result('EdgeBetweennessPartition', part, params={'k': k}, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('EdgeBetweennessPartition failed or exhausted:', e)

    # 3) Edge Current Flow Betweenness Partition - similar
    try:
        # Testa vários valores de k, que representa o número desejado de comunidades a serem formadas.
        for k in range(2, min(max_k_try+1, n)):
            # Executa o algoritmo com vários tamanhos de k
            part = edge_current_flow_betweenness_partition(GF, k, weight=weight)
            # Salva os resultados do algoritmo
            save_result('EdgeCurrentFlowBetweennessPartition', part, params={'k': k}, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('EdgeCurrentFlowBetweennessPartition failed or exhausted:', e)

    # 4) K-clique communities - try k=3..6
    try:
        # Testa vários valores de k entre 3 e 6, que representa o número desejado de comunidades a serem formadas.
        for k in range(3, min(7, max_k_try+3)):
            # Executa o algoritmo com vários tamanhos de k
            comms = list(k_clique_communities(GF, k))
            # Caso tenha resultado salva
            if comms:
                # Salva os resultados do algoritmo
                save_result('K-clique', comms, params={'k': k}, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('K-clique failed:', e)

    # 5) Greedy Modularity Communities (Clauset-Newman-Moore)
    try:
        # Executa o algoritmo
        comms = list(greedy_modularity_communities(GF, weight=weight))
        # Salva os resultados do algoritmo
        save_result('GreedyModularity', comms, params=None, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('GreedyModularity failed:', e)

    # 6) Naive Greedy Modularity Communities (reference O(n^4) implementation)
    # try:
    #     comms = list(naive_greedy_modularity_communities(G, weight=weight))
    #     save_result('NaiveGreedyModularity', comms, params=None)
    # except Exception as e:
    #     print('NaiveGreedyModularity failed:', e)

    # 7) Lukes Partitioning - apenas para árvores (conectadas, acíclicas)
    try:
        if nx.is_tree(GF):
            # O parâmetro `max_size` deve ser fornecido: tente alguns valores razoáveis.
            for max_size in [1, 2, 5, 10, max(1, n//4)]:
                # Executa o algoritmo com vários tamanhos máximos
                part = list(lukes_partitioning(GF, max_size))
                # Salva os resultados do algoritmo
                save_result('LukesPartitioning', part, params={'max_size': max_size}, weight=weight, node_to_group=node_to_group)
        else:
            print('LukesPartitioning skipped (graph is not a tree).')
    except Exception as e:
        print('LukesPartitioning failed:', e)

    # 8) Label Propagation (synchronous) - networkx.label_propagation_communities
    try:
        # Executa o algoritmo
        comms = list(label_propagation_communities(GF))
        # Salva os resultados do algoritmo
        save_result('LabelPropagation', comms, params=None, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('LabelPropagation failed:', e)

    # 9) Fast/Asynchronous Label Propagation (asyn_lpa_communities)
    try:
        # Retorna algo semelhante a um gerador; tente o padrão e com semente.
        comms = list(asyn_lpa_communities(GF, weight=weight, seed=42))
        # Salva os resultados do algoritmo
        save_result('AsynchronousLabelPropagation', comms, params={'seed': 42}, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('AsynchronousLabelPropagation failed:', e)

    # 10) Louvain Communities
    try:
        # Executa o algoritmo 
        comms = list(louvain_communities(GF, weight=weight))
        # Salva os resultados do algoritmo
        save_result('Louvain', comms, params=None, weight=weight, node_to_group=node_to_group)
    except Exception as e:
        print('Louvain failed:', e)

    # 11) Asynchronous Fluid Communities (asyn_fluidc) - precisa de k (número de comunidades)
    try:
        # Requer um grafo conectado
        if not nx.is_connected(GF):
            # Aplique a cada componente conectado separadamente e combine os resultados.
            for comp in nx.connected_components(GF):
                # Pega um subgrafo conectado
                H = GF.subgraph(comp).copy()
                # Pega o número de nós do subgrafo
                comp_n = H.number_of_nodes()
                # Calcula o número máximo de interações
                max_k = min(max_k_try, max(2, comp_n//2))
                # Loop para executar o máximo de interações
                for k in range(2, max_k+1):
                    try:
                        # Executa o algoritmo
                        comms = list(asyn_fluidc(H, k, max_iter=200, seed=42))
                        # asyn_fluidc retorna um gerador de conjuntos; precisamos fazer a junção em uma partição global.
                        # Construir uma partição global onde os nós fora de `comp` sejam conjuntos unitários.
                        partition = [set(c) for c in comms]
                        # Adicionar comunidades singleton para nós que não estão em comp
                        outside = set(GF.nodes()) - set(H.nodes())
                        partition.extend([{n} for n in outside])
                        # Salva os resultados do algoritmo
                        save_result('AsynchronousFluidC', partition, params={'k': k, 'component_size': comp_n}, weight=weight, node_to_group=node_to_group)
                    except Exception:
                        pass
        else:
            # Conectado
            # Calcula o número máximo de interações
            max_k = min(max_k_try, max(2, n//2))
             # Loop para executar o máximo de interações
            for k in range(2, max_k+1):
                try:
                    # Executa o algoritmo 
                    comms = list(asyn_fluidc(GF, k, max_iter=200, seed=42))
                    # Salva os resultados do algoritmo
                    save_result('AsynchronousFluidC', comms, params={'k': k}, weight=weight, node_to_group=node_to_group)
                except Exception:
                    pass
    except Exception as e:
        print('AsynchronousFluidC failed:', e)

    # Resuldado
    df = pd.DataFrame(results)

    # Classificação dos resultados: primária por modularidade (em ordem decrescente), secundária por desempenho.
    df_sorted = df.sort_values(by=['modularity', 'performance'], ascending=[False, False])

    return df_sorted.reset_index(drop=True), partitions
```
